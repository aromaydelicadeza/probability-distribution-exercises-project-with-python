# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

## Ejercicio 1

Escriba las funciones anteriores pero para distribuciones normales y uniformes.

```python
#normal

from scipy.stats import norm

def dnormal(x, mean, std):
    """
    Calcula la estimación puntual de la distribución normal.
    """
    result = norm.pdf(x = x, loc=mean, scale = std)

    return result

def pnormal(q, mean, std):
    """
    Calcula el acumulado de la distribución normal.
    """
    result = norm.cdf(x = q, loc = mean, scale = std)

    return result

def qnormal(p, mean, std):

    """
    Calcula la fucnión cuantil a partir de la distribución normal.
    """
    result = norm.ppf(q = p, loc = mean, scale = std)

    return result

def rnormal(size, mean, std):
    """
    Genera variables aleatorias a partir de la distribución normal.
    """
    result = norm.rvs(size = size, loc = mean, scale = std)

    return result
```

```python
print(f"Probabilidad que una Normal(0, 1) tome el valor 2: {dnormal(2, 0,1)}")
print(f"Probabilidad que una Normal(0, 1) tome un valor menor que 2: {pnormal(2, 0,1)}")
print(f"Qué valor de una Normal(0, 1) tiene una probabilidad acumulativa de 0.9? {qnormal(0.9, 0,1)}")
print(f"Genera 2000 valores aleatorios desde una distribución Normal(0, 1): {rnormal(2000, 0,1)}")
```

```python
mean = 0
std = 1
data = rnormal(20000, mean, std)

pdf = dnormal(data, mean,std)
cdf = pnormal(data, mean,std)
ppf = qnormal(data, mean,std)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pdf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

plt.tight_layout()
plt.show()
```

```python
#uniforme

from scipy.stats import uniform

def duniform(x, low, high):
    """
    Calcula la estimación puntual de la distribución normal.
    """
    result = uniform.pdf(x = x, loc=low, scale = (high-low))

    return result

def puniform(q, low, high):
    """
    Calcula el acumulado de la distribución normal.
    """
    result = uniform.cdf(x = q, loc = low, scale = (high-low))

    return result

def quniform(p, low, high):
    """
    Calcula la función cuantil a partir de la distribución normal.
    """
    result = uniform.ppf(q = p, loc = low, scale = (high-low))

    return result

def runiform(size, low, high):
    """
    Genera variables aleatorias a partir de la distribución normal.
    """
    result = uniform.rvs(size = size, loc = low, scale = (high-low))

    return result




```

```python
print(f"Probabilidad que una Uniforme(0, 1) tome el valor 0.5: {duniform(0.5, 0,1)}")
print(f"Probabilidad que una Uniforme(0, 1) tome un valor menor que 0.8: {puniform(0.8, 0,1)}")
print(f"Qué valor de una Uniforme(0, 1) tiene una probabilidad acumulativa de 0.9? {quniform(0.9, 0,1)}")
print(f"Genera 2000 valores aleatorios desde una distribución Uniforme(0, 1): {runiform(2000, 0,1)}")
```

```python
low = 0
high = 1
data = runiform(20000, low, high)

pdf = duniform(data, low, high)
cdf = puniform(data, low, high)
ppf = quniform(data, low, high)

fig, axis = plt.subplots(3, 1, figsize = (5,7))

axis[0].stem(data, pdf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

plt.tight_layout()
plt.show()
```
